```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

sample = 50
mean1 = [5]
mean2 = [10]
variance = 0.5

class_0 = np.random.normal(loc=mean1, scale=np.sqrt(variance), size=(sample, 2))
class_1 = np.random.normal(loc=mean2, scale=np.sqrt(variance), size=(sample, 2))

X = np.vstack((class_0, class_1))
Y = np.array([0] * sample + [1] * sample)

data = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])
data['Class'] = Y 

data.to_csv('data.csv', index=False)

plt.scatter(X[Y == 0][:, 0], X[Y == 0][:, 1], color='red')
plt.scatter(X[Y == 1][:, 0], X[Y == 1][:, 1], color='green')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
